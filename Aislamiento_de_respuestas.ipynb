##Reorden

```
import pandas as pd
import numpy as np
dfEvaluar = pd.read_csv('asteroidEval.csv')
for columna in dfEvaluar.columns:
  if (columna != "full_name"): dfEvaluar = dfEvaluar.drop(columna, 1)
dfOriginal = dfEvaluar.copy()
dfOriginal["full_name"] = pd.read_csv('orden a corregir.txt', delimiter = "\n")
dfOriginal['danger'] = pd.read_csv('diametros a corregir.txt', delimiter = "\n")#pd.read_csv('respuestas a corregir.txt', delimiter = "\n")
dfOriginal2 = dfOriginal.copy()
dfEvaluar['danger'] = [0]*len(dfEvaluar)
print(dfEvaluar,"\n\n\n",dfOriginal,"\n\n\n",dfOriginal2)
```

```
#from google.colab import output
for indice1 in range(len(dfEvaluar)):
  for indice2 in range(len(dfOriginal2)):
    #output.clear()
    #print(indice1, indice2)
    if ((dfEvaluar['full_name'].iloc[indice1]) == (dfOriginal2['full_name'].iloc[indice2])):
      #print(indice1, len(dfEvaluar), indice2, len(dfOriginal2)) #Este es útil
      dfEvaluar['danger'].iloc[indice1] = dfOriginal2['danger'].iloc[indice2]
      dfOriginal2 = dfOriginal2.drop(indice2).reset_index(drop=True)
      break

print(dfEvaluar,"\n\n\n",dfOriginal,"\n\n\n",dfOriginal2)
```

```
with open('Diametros_reales.txt', 'wb') as f:#with open('Respuestas_reales.txt', 'wb') as f:
  for pred in list(dfEvaluar.iloc[:,1]): 
    f.write((str(pred)+'\n').encode())
```
